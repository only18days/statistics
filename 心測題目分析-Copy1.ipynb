
### Mean imputation

```python
df = pd.DataFrame({'Q16':Q16, 'Q17':Q17, 'Q18':Q18, 'Q19':Q19, 'Q20':Q20})
df
```

```python
imputer = SimpleImputer(strategy='mean', missing_values = np.nan)
imputer = imputer.fit(df[['Q16','Q17','Q18','Q19','Q20']])
df[['Q16','Q17','Q18','Q19','Q20']] = imputer.transform(df[['Q16','Q17','Q18','Q19','Q20']])
df
```

```python
# Redefine Q16-Q20 (as series)
Q16 = df[['Q16']].squeeze()
Q17 = df[['Q17']].squeeze()
Q18 = df[['Q18']].squeeze()
Q19 = df[['Q19']].squeeze()
Q20 = df[['Q20']].squeeze()
```
